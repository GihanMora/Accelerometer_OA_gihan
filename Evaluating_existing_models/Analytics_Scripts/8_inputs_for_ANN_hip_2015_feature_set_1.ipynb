```
parameter_files = [f for f in listdir(Actigraph_30s_hip_parameters) if isfile(join(Actigraph_30s_hip_parameters, f))]
print(parameter_files)
```

```
for f in tqdm(parameter_files):
  try:
    fe_data = pd.read_csv(join(Actigraph_30s_hip_parameters, f))
    # print(fe_data.head())
    columns = ['Participant', 
              'Activity',
            'OC_METs', 
            'TimeSec', 
            'Mean_AGhipX', 
            'Mean_AGhipY', 
            'Mean_AGhipZ', 
            'Var_AGhipX', 
            'Var_AGhipY', 
            'Var_AGhipZ',]
    result_data = pd.DataFrame(columns=columns)
    result_data['Participant'] = [1]*len(fe_data)
    result_data['Activity'] = ['test_activity']*len(fe_data)
    result_data['OC_METs'] = [10]*len(fe_data)
    result_data['TimeSec'] = [100]*len(fe_data)
    result_data['Mean_AGhipX'] = fe_data['raw_XMean']
    result_data['Mean_AGhipY'] = fe_data['raw_YMean']
    result_data['Mean_AGhipZ'] = fe_data['raw_ZMean']
    result_data['Var_AGhipX'] = fe_data['raw_XVar']
    result_data['Var_AGhipY'] = fe_data['raw_YVar']
    result_data['Var_AGhipZ'] = fe_data['raw_ZVar']

    print(result_data.head())
    
    output_f = f.replace('.csv', '_ANN_input_hip_2015_feature_set_2.txt')
    output_filename = join(Actigraph_output_folder, output_f)

    result_data.to_csv(output_filename, sep='\t', index=False, line_terminator='\n')

  except Exception as e:
    f = open('errors.txt','w+')
    f.write(" "+str(e)+" \n")
    f.close()
```
