```
from os import listdir
from os.path import isfile, join
import numpy as np
import pandas as pd
import sys, time
import matplotlib.pyplot as plt
```

```
epoch_folder = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/Epoch30/'
predictions_folder_1 = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/outputs_for_ANN_thigh_2015_feature_set_3/'
predictions_folder_2 = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/outputs_for_ANN_thigh_2015_feature_set_2/'
predictions_folder_3 = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/outputs_for_ANN_thigh_2015_feature_set_1/'
predictions_folder_4 = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/outputs_for_ANN_thigh_2015_feature_set_4/'

output_folder = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/montoye_thigh/combined_results_ANN_thigh_2015/'

epoch_data_files = [f for f in listdir(epoch_folder) if isfile(join(epoch_folder, f))]
prediction_data_files_1 = [f for f in listdir(predictions_folder_1) if isfile(join(predictions_folder_1, f))]
prediction_data_files_2 = [f for f in listdir(predictions_folder_2) if isfile(join(predictions_folder_2, f))]
prediction_data_files_3 = [f for f in listdir(predictions_folder_3) if isfile(join(predictions_folder_3, f))]
prediction_data_files_4 = [f for f in listdir(predictions_folder_4) if isfile(join(predictions_folder_4, f))]
print([len(epoch_data_files),len(prediction_data_files_1),len(prediction_data_files_2),len(prediction_data_files_3),len(prediction_data_files_4)])

if len(epoch_data_files) == len(prediction_data_files_1) == len(prediction_data_files_2) == len(prediction_data_files_3) == len(prediction_data_files_4):
    for epoch_file, pred_file_1, pred_file_2,pred_file_3, pred_file_4 in zip(sorted(epoch_data_files), sorted(prediction_data_files_1), sorted(prediction_data_files_2), sorted(prediction_data_files_3), sorted(prediction_data_files_4)):
      if(epoch_file.split('.csv')[0]==pred_file_1.split('_predicted_thigh_EE_2015_feature_set_3.csv')[0]==pred_file_2.split('_predicted_thigh_EE_2015_feature_set_2.csv')[0]==pred_file_3.split('_ANN_input_hip_2015_feature_set_1.txt_predicted_thigh_EE_2015_feature_set_1.csv')[0]==pred_file_4.split('_ANN_input_hip_2015_feature_set_4.txt_predicted_thigh_EE_2015_feature_set_4.csv')[0]):
        print(epoch_file,pred_file_1,pred_file_2,pred_file_3,pred_file_4)
        result_data = pd.read_csv(epoch_folder + epoch_file)
        predict_data_1 = pd.read_csv(predictions_folder_1 + pred_file_1)
        predict_data_2 = pd.read_csv(predictions_folder_2 + pred_file_2)
        predict_data_3 = pd.read_csv(predictions_folder_3 + pred_file_3)
        predict_data_4 = pd.read_csv(predictions_folder_4 + pred_file_4)

        result_data['predicted_ee_montoye_2015_feature_set_3'] = predict_data_1['V1']
        result_data['predicted_ee_montoye_2015_feature_set_2'] = predict_data_2['V1']
        result_data['predicted_ee_montoye_2015_feature_set_1'] = predict_data_3['V1']
        result_data['predicted_ee_montoye_2015_feature_set_4'] = predict_data_4['V1']
        # print(result_data.head())
        # print(predict_data['V1'])
        result_data.to_csv((output_folder+pred_file_1.split('_predicted_thigh_EE_2015_feature_set_3.csv')[0]+'_ANN_2015_thigh_combined.csv'), sep=',')
```
