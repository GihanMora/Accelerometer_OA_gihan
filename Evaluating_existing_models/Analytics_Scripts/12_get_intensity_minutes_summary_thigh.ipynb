```
import csv
excel_folder = '/content/drive/My Drive/OA activity data_La Trobe/Analytics_Scripts/Gihan_analysis/processed data/thigh_minute_data/'

def get_res_cat(name,df):
  results = []
  for k in ['1','2','3']:
    try:
      value = (str(round(df[k].mean(),3))+' ± '+str(round(df[k].std(),3))).strip()
      results_temp = [name+'_'+k,len(df[k]),round(df[k].mean(),3),round(df[k].std(),3),value]
      results.append(results_temp)
    except Exception:
      continue
  return results

def get_results_for_book(book_name,model_name):
  all = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='all')
  female = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='female')
  male = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='male')
  obese = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='obese')
  overweight = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='overweight')
  normal = pd.read_excel(open(excel_folder+book_name, 'rb'),sheet_name='normal')
  results = []
  results.append(get_res_cat(model_name+'_all',all))
  results.append(get_res_cat(model_name+'_female',female))
  results.append(get_res_cat(model_name+'_male',male))
  results.append(get_res_cat(model_name+'_obese',obese))
  results.append(get_res_cat(model_name+'_overweight',overweight))
  results.append(get_res_cat(model_name+'_normal',normal))

  return results

all_results = []
all_results.append(get_results_for_book('1_thigh_reference_minute_values.xlsx','reference_thigh'))
all_results.append(get_results_for_book('4_thigh_mont_1_minute_values.xlsx','mont_1_thigh'))
all_results.append(get_results_for_book('4_thigh_mont_2_minute_values.xlsx','mont_2_thigh'))
all_results.append(get_results_for_book('4_thigh_mont_3_minute_values.xlsx','mont_3_thigh'))
all_results.append(get_results_for_book('4_thigh_mont_4_minute_values.xlsx','mont_4_thigh'))
all_results.append(get_results_for_book('4_thigh_mont_17_minute_values.xlsx','mont_17_thigh'))

all_results.append(get_results_for_book('4_thigh_white_1_minute_values.xlsx','white_1_thigh'))
all_results.append(get_results_for_book('4_thigh_white_2_minute_values.xlsx','white_2_thigh'))


results_file = excel_folder+'5_summary_thigh.csv'
f = open(results_file, 'w', encoding='utf-8')
writer = csv.writer(f)
writer.writerow(['model','count','mean','standard_deviation','value'])
for l in all_results:
  for ele in l:
    for el in ele:
      print(el)
      writer.writerow(el)

f.close()
```

```
def categorize(input_name,output_sheet_name):
  all_df = pd.read_csv(join(ActiPAL_output_folder,input_name))

  female = all_df['sex'] == 'female'
  male = all_df['sex'] == 'male'
  obese = all_df['BMI_range'] == 'obese'
  overweight = all_df['BMI_range'] == 'overweight'
  normal = all_df['BMI_range'] == 'normal'


  male_df = all_df[male]
  female_df = all_df[female]
  obese_df = all_df[obese]
  overweight_df = all_df[overweight]
  normal_df = all_df[normal]

  # print(['all,male,female',len(all_df),len(male_df),len(female_df)])
  # print(['all,male,female',all_df['1'].mean,len(male_df),len(female_df)])

  with pd.ExcelWriter(join(ActiPAL_output_folder,output_sheet_name)) as writer: 
      all_df.to_excel(writer, sheet_name='all')
      female_df.to_excel(writer, sheet_name='female')
      male_df.to_excel(writer, sheet_name='male')
      obese_df.to_excel(writer, sheet_name='obese')
      overweight_df.to_excel(writer, sheet_name='overweight')
      normal_df.to_excel(writer, sheet_name='normal')
```
